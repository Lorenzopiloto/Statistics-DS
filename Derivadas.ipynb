# Derivadas

# Módulos

```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
```

- A função diff facilita a resolução

```python
smp.diff(3)
```

```python
t = 2
smp.diff(t)
```

- Precisamo definir a variável como símbolo e depois extrair a derivada

```python
x = smp.Symbol('x')
```

```python
f = x**2
smp.diff(f)
```

- Vamos gerar um gráfico a partir das funções

```python
funcao_original = x**2
funcao_derivada = smp.diff(funcao_original)
```

```python
funcao_original, funcao_derivada
```

- A função lambdify ajuda a transformar numa função em que podemos utilizar em operações

```python
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
```

```python
y = [funcao_original(i) for i in range(-100, 100)]
plt.plot(y)
```

```python
y_d = [funcao_derivada(i) for i in range(-100, 100)]
plt.plot(y_d)
```
